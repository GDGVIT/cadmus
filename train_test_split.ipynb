### Reading the dataset

```python
df=pd.read_csv('data.csv',index_col ="gloss_split")
df.head()
```

### Splitting into test and train datasets

```python
test = pd.DataFrame(df.loc[["test", "val"]]) 
train = pd.DataFrame(df.loc[['train']])
```

```python
test.reset_index(drop=True, inplace=True)
train.reset_index(drop=True, inplace=True)
```

```python
test=test.drop(test.columns[[0]],axis = 1) 
train=train.drop(train.columns[[0]],axis = 1) 
```

```python
import cv2
vidcap = cv2.VideoCapture('big_buck_bunny_720p_5mb.mp4')
success,image = vidcap.read()
count = 0
while success:
  cv2.imwrite("frame%d.jpg" % count, image)     # save frame as JPEG file      
  success,image = vidcap.read()
  print('Read a new frame: ', success)
  count += 1
```

```python
for i in train['gloss_video_id']:
    path = "/home/sharanya/Desktop/cadmus/train/"+str(i)+"frame"
    os.mkdir(path);
    vidcap = cv2.VideoCapture('videos/'+str(i)+'.mp4')   # capturing the video from the given path
    success,image = vidcap.read()
    count = 0
    
    while success:
      filename = str(i)+"_"+str(count)+'.jpg'
      cv2.imwrite(os.path.join(path , filename), image)     # save frame as JPEG file      
      success,image = vidcap.read()
      print('Read a new frame: ', success)
      count += 1
```
